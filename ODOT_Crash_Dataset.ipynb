```python
import pandas as pd
import numpy as np
```

```python
col_dtype = {"Crash ID": int, "Crash Year": str, "Driver License Status": str }
```

```python
crash_marion = pd.read_csv("Marion.csv", dtype = col_dtype)
crash_clackamas = pd.read_csv("Clackamas.csv", dtype = col_dtype)
#crash_washington = pd.read_csv("Washington.csv")
#crash_multnomah = pd.read_csv("multnomah.csv")
code = pd.read_csv("CDS Code Translate.csv")
```

```python
# Return the code for Driver License Status
code1 = code[["PARTIC.DRVR_LIC_STAT_CD", "Driver License Status", "PARTIC.DRVR_RES_STAT_CD", "Residence of Driver"]]
code1.iloc[0:7]
```

```python
# Union all 4 counties
crash = crash_marion.append([crash_clackamas])
print crash["Participant ID"].dtype
crash.head()
```

```python
# Return a DF with only the relevant columns
crash_select1 = crash[[ "Crash ID", "Participant ID", "Driver License Status", "Driver Residence"]].dropna(subset = ["Participant ID"])
crash_select2 = crash[["Crash ID", "Crash Year"]].dropna()
crash_select = crash_select1.merge(crash_select2, on = "Crash ID", how = "inner")
crash_select.head()
```

```python
crash_select["Driver License Status"].value_counts()
```

```python
# Select rows with valid in-state and valid out-of-state licenses
crash_select = crash_select.loc[(crash_select["Driver License Status"] == "1") | (crash_select["Driver License Status"] == "2")]
```

```python
crash_select.head()
```

```python
# Return the number of crasher per year
crash_select.groupby("Crash Year")["Crash ID"].count()
```

```python
# Return the distribution of driver licenses by crash year
crash_select.groupby(["Crash Year", "Driver License Status"]).count()
```
